```python

# Start Setting Up User Interface!!!

pokeSimulation = True # activates the program
print("PROFESSOR ROWAN:")
print("    Hello there!")
print("    It's so very nice to meet you!")
print("    Welcome to the world of Pokémon!")
print("    My name is Rowan.")
print("    However, everyone just calls me the Pokémon Professor.")
print("    Before I go any further, is this your first adventure?")
print("    If you need advice, I'm certainly capable of giving it.")
print("")
    
    
while pokeSimulation == True:
    
    # prompt the user of the choices
    print("    (A) Find info on Pokémon by Name")
    print("    (Y) Find info on Pokémon by Type")
    print("    (B) Find info on Pokémon by Ndex (National Index #)")
    print("    (X) No Info Needed! \n")
    
    userSearch = input("Me: ")
    
     # program carries out function based on the user's input
    if userSearch == 'A':
        print("PROFESSOR ROWAN: Ah I see, what's the name of the Pokémon you're curious about?")
        userSearch = input("Pokémon Name: ")
        findWithName(userSearch)
        
    elif userSearch == 'Y':
        print("PROFESSOR ROWAN: Ah I see, what's the minimum amount of types that the Pokémon you're curious about have?")
        userSearch = input("Minimum # of Types in Pokémon (1 or 2): ")
        print("\nPROFESSOR ROWAN: Marvelous! So you're interested in finding Pokémon with " + userSearch + " types!\n")
        if userSearch == str(1):
            print("What type of Pokémon are you curious about?")
            userSearch1 = input("Type of Pokémon: ")
            findWithType(userSearch1)
        elif userSearch == str(2):
            print("What is one of the types for the Pokémon you're curious about?")
            userSearch1 = input("Type #1 of Pokémon: ")
            print("And what is second of the types for the Pokémon you're curious about?")
            userSearch2 = input("Type #2 of Pokémon: ")
            findWithType(userSearch1, userSearch2)
        else:
            print("Hmm, wait just a moment. I don't believe we have Pokémon with that number of types in our database.")
        
    elif userSearch == 'B':
        print("PROFESSOR ROWAN: Ah I see, what's the official National Index of the Pokémon you're curious about?")
        userSearch = input("Pokémon Ndex: ")
        findWithNdex(userSearch)
        
    elif userSearch == 'X':
        pokeSimulation = False
        print("PROFESSOR ROWAN:")
        print("    All right, the time has come.")
        print("    Your very own tale of grand adventure is about to unfold.")
        print("    On your journey, you will meet countless Pokémon and people.")
        print("    I'm sure that along the way you will discover many things, perhaps even something about yourself.")
        print("    Now, go on, leap into the world of Pokémon!")
        print("*Professor Rowan dramatically disappears and Pokémon Diamond & Pearl soundtrack begins playing...*")
        break
        
    else:
        print("PROFESSOR ROWAN: Pardon? I'm an old man, mind repeating yourself again?")
    
    
    # prompt user again for next looping
    print("\n--------------------------------------------------")
    print("PROFESSOR ROWAN: Would you like to know more about anything else?")
    
   
```
